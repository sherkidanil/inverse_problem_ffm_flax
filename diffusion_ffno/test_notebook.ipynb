# 1D problem, single target

## Dataset

```python
Burgers_data = jnp.load("/mnt/local/dataset/by-domain/pde/PDE_datasets/PDE_bench/1D_Burgers_Sols_Nu0.001.npz")

Burgers_features = Burgers_data['features'][:, :, :, 0]
Burgers_targets = Burgers_data['targets'][:, :, :, 20]
Burgers_coordinates = Burgers_data['coordinates'][:1, :, 0]

Burgers_features = Burgers_features / jnp.mean(jnp.linalg.norm(Burgers_features, ord=jnp.inf, axis=2, keepdims=True), axis=0, keepdims=True)
Burgers_targets = Burgers_targets / jnp.mean(jnp.linalg.norm(Burgers_targets, ord=jnp.inf, axis=2, keepdims=True), axis=0, keepdims=True)

print("features", Burgers_features.shape)
print("targets", Burgers_targets.shape)
print("coordinates", Burgers_coordinates.shape)

del Burgers_data
```

```python
fig, ax = plt.subplots(1, 2, figsize=(9, 3))
for s in [13, 45, 667, 22]:
  ax[0].plot(Burgers_coordinates[0], Burgers_features[s, 0])
  ax[1].plot(Burgers_coordinates[0], Burgers_targets[s, 0])

ax[0].set_title("features")
ax[1].set_title("targets");
```

## FFNO

```python
from baselines import evaluate as eval_b, train as train_b
from baselines.architectures import ffno as ffno_b
```

```python
D = 1
learning_rate = 1e-4
N_processor = 32
N_train = 4000
N_run = 20000
N_batch = 20
N_layers = 4
N_modes = 16
N_drop = N_run // 2
gamma = 0.5

key = random.PRNGKey(11)
keys = random.split(key, 2)
N_features = [Burgers_coordinates.shape[0] + Burgers_features.shape[1], N_processor, Burgers_targets.shape[1]]
model = ffno_b.FFNO(N_layers, N_features, N_modes, D, keys[0])

learning_rate = optax.exponential_decay(learning_rate, N_drop, gamma)
optim = optax.lion(learning_rate=learning_rate)
opt_state = optim.init(eqx.filter(model, eqx.is_array))

ind = jnp.arange(Burgers_features.shape[0])
ind_train, ind_test = ind[:N_train], ind[N_train:]
n = random.choice(keys[1], ind_train, shape = (N_run, N_batch))
carry = [model, Burgers_features, Burgers_targets, Burgers_coordinates, opt_state]

make_step_scan_ = lambda a, b: train_b.make_step_scan(a, b, optim)
res, losses = scan(make_step_scan_, carry, n)
model = res[0]

plt.yscale("log")
plt.plot(losses)

_, train_errors = scan(eval_b.compute_error_scan, [model, Burgers_features, Burgers_targets, Burgers_coordinates], ind_train)
_, test_errors = scan(eval_b.compute_error_scan, [model, Burgers_features, Burgers_targets, Burgers_coordinates], ind_test)

mean_train_error = jnp.mean(train_errors, axis=0)
mean_test_error = jnp.mean(test_errors, axis=0)

print("train errors", mean_train_error)
print("test errors", mean_test_error);
```

## Flow matching + FFNO

```python
from flow_matching import train as train_fm, priors, flows
from flow_matching.architectures import ffno as ffno_fm

from flow_matching.integrators import explicit_Euler, integrator
from flow_matching.evaluate import get_statistics_scan, compute_error as compute_error_fm
```

```python
D = 1
learning_rate = 1e-3
N_processor = 32
N_train = 4000
N_run = 50000
N_batch = 100
N_layers = 4
N_modes = 16
N_drop = N_run // 4
gamma = 0.5
scale = 0.001
po = 2.0
N = 150

key = random.PRNGKey(11)
keys = random.split(key, 3)
N_features = [Burgers_coordinates.shape[0] + Burgers_features.shape[1] + Burgers_targets.shape[1] + 1, N_processor, Burgers_targets.shape[1]]
model = ffno_fm.flow_FFNO(N_layers, N_features, N_modes, D, keys[0])

learning_rate = optax.exponential_decay(learning_rate, N_drop, gamma)
optim = optax.lion(learning_rate=learning_rate)
opt_state = optim.init(eqx.filter(model, eqx.is_array))

ind = jnp.arange(Burgers_features.shape[0])
ind_train, ind_test = ind[:N_train], ind[N_train:]
n = random.choice(keys[1], ind_train, shape = (N_run, N_batch))

ind = jnp.arange(Burgers_features.shape[0])
ind_train, ind_test = ind[:N_train], ind[N_train:]
n = random.choice(keys[1], ind_train, shape = (N_run, N_batch))

carry = [model, Burgers_targets, Burgers_features, Burgers_coordinates, opt_state, keys[2]]

flow_params = [0.0, ]
flow = lambda target_1, target_0, t: flows.optimal_transport(target_1, target_0, t, flow_params)

basis, freq = priors.get_basis_normal_periodic(Burgers_coordinates, N)
prior_params = [basis, freq, scale, po]
prior = lambda key: priors.normal_periodic(key, prior_params)

make_step_scan_ = lambda a, b: train_fm.make_step_scan(a, b, optim, flow, prior)
carry, losses = scan(make_step_scan_, carry, n)
model = carry[0]

plt.yscale("log")
plt.plot(losses);
```

```python
N_t = 50
N_samples = 50
t = jnp.linspace(0, 1, N_t)
dt = t[1] - t[0]
integrator_ = lambda carry, t: integrator(carry, t, explicit_Euler)

get_statistics_scan_ = lambda carry, ind: get_statistics_scan(carry, ind, N_samples, integrator_, prior)
_, train_predictions = scan(get_statistics_scan_, [model, Burgers_targets, Burgers_features, Burgers_coordinates, t, dt, keys[3]], ind_train)
_, test_predictions = scan(get_statistics_scan_, [model, Burgers_targets, Burgers_features, Burgers_coordinates, t, dt, keys[3]], ind_test)
```

```python
train_errors = compute_error_fm(Burgers_targets[ind_train], train_predictions[:, 0])
test_errors = compute_error_fm(Burgers_targets[ind_test], test_predictions[:, 0])

mean_train_error = jnp.mean(train_errors, axis=0)
mean_test_error = jnp.mean(test_errors, axis=0)

print("train errors", mean_train_error)
print("test errors", mean_test_error);
```

```python
train_correlations = vmap(jnp.corrcoef, in_axes=(0, 0))(jnp.abs(train_predictions[:, 0, 0] - Burgers_targets[ind_train][:, 0]), train_predictions[:, 1, 0])[:, 0, 1]
test_correlations = vmap(jnp.corrcoef, in_axes=(0, 0))(jnp.abs(test_predictions[:, 0, 0] - Burgers_targets[ind_test][:, 0]), test_predictions[:, 1, 0])[:, 0, 1]

mean_train_correlation = jnp.mean(train_correlations, axis=0)
mean_test_correlation = jnp.mean(test_correlations, axis=0)

print("train correlation", mean_train_correlation )
print("test correlation", mean_test_correlation);
```
