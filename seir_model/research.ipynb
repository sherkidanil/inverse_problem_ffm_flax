```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss_with_noise = np.load('models/seir_4p_ds_without_noise/losses.npy')
loss_without_noise = np.load('models/seir_4p_ds_with_noise/losses.npy')
```

```python
plt.figure(figsize=(12,6), dpi=300)
plt.plot(loss_with_noise, label=fr'$x_t$ with noise, err: $2.23\% \pm 0.15\%$')
plt.plot(loss_with_noise, label=fr'$x_t$ without noise, err: $0.16\% \pm 0.10\%$')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
# plt.yscale('log')
plt.show()
```

```python
from scipy.ndimage import gaussian_filter1d

plt.figure(figsize=(12,6), dpi=300)
plt.plot(gaussian_filter1d(loss_with_noise, 1000), label=fr'$x_t$ with noise, err: $2.2\% \pm 0.15\%$')
plt.plot(gaussian_filter1d(loss_without_noise, 1000), label=fr'$x_t$ without noise, err: $0.16\% \pm 0.10\%$')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.yscale('log')
# plt.grid(True)
plt.show()
```
